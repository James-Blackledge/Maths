```python
#Univariate Data:

#The histogram below is skewed. This happens when the mean and the median of the data are not equal.

#If the histogram is not skewed, then we can say that it is symmetric.
#Data is symmetric when the mean and median are equal.
#If the graph is shaped like a bell curve, the peak will contain the mean, median and mode.

#An average (Mean, Median, Mode) can be described as 'Measures of location (Central Tendency)' - SUMMARY STATISTICS

#'Measures of Variation (of Dispersion)' - IQR, Range, Standard Deviation.

#IQR identifies the middle spread of data. Not as affected by anomalies.
```

```python
from random import randint
```

```python
def snake():
    running = True; a = 0; b = 0; total = 0; rolls = 1
    
    while running:
        a = randint(1, 6); b = randint(1, 6)

        if not (1 in [a, b]):
            total += (a + b); rolls += 1
        else:
            running = False
        
    return (rolls, total)
```

```python
def mean(data_list):
    return sum(data_list) / len(data_list)
```

```python
def median(data_list):
    data_list = sorted(data_list)
    
    middle_odd = data_list[len(data_list) // 2]
    middle_even = sum(data_list[(middle_odd - 2) : middle_odd]) / 2
    
    return (middle_odd, middle_even)[len(data_list) % 2 == 0]
```

```python
def mode(data_list):
    data_list = sorted(data_list)
    
    a = [i for i in set(data_list)]
    b = [data_list.count(i) for i in a]
    
    zipped_dict = dict(zip(b, a))
    
    return zipped_dict[max(zipped_dict.keys())]
```

```python
def xrange(data_list):
    data_list = sorted(data_list)
    
    return data_list[-1] - data_list[0]
```

```python
data = [snake()[0] for i in range(1000)]
```

```python
mean(data)
```

```python
median(data)
```

```python
mode(data)
```

```python
xrange(data)
```
