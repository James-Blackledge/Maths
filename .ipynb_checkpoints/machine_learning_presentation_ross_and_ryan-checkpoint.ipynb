```python
import random, numpy as np, matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = float(random.randint(0, 10))

def feed_forward(x):
    return x * w

def calculate_loss(x, y):
    return (feed_forward(x) - y) ** 2

def compute_gradient(x, y):
    return 2 * x * ((x * w) - y)

alpha = 0.01

for epoch in range(1, 100):
    for x, y in zip(x_data, y_data):
        gradient = compute_gradient(x, y)
        
        w -= alpha * gradient
        loss = calculate_loss(x, y)
        
        print("X: {0}, Y: {1}, Gradient: {2}".format(x, y, gradient))
    print("Epoch: {0}, Weight: {1}, Los: {2}".format(epoch, w, loss))
```

## $\hat{y} = x * w$
## $loss = (\hat{y} - y)^2$
## $w = w - \alpha\frac{\delta loss}{\delta w}$

### Chain Rule

```python
def feed_forward(x, w):
    return x * w
```

```python
def loss(x, y):
    return [[(feed_forward(a, i) - b) ** 2 for a, b in zip(x, y)] for i in np.arange(0.0, 4.1, 0.1)]

plt.plot(np.arange(0.0, 4.1, 0.1), loss(x_data, y_data))
```
